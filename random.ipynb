```python
dfs = [tuition_cost, tuition_income, salary_potential, diversity_school]

all_schools = {}

for df in dfs:
    for _, row in df.iterrows():
        school = row['name']
        if school not in all_schools.keys():
            all_schools[school] = 1
        else:
            all_schools[school] += 1

len(all_schools.keys())
```

```python
counts = all_schools.values()
sorted_counts = sorted(counts)
med = np.median(sorted_counts)
q1 = np.quantile(sorted_counts, 0.25)
q3 = np.quantile(sorted_counts, 0.75)
iqr = q3 - q1
box = 1.5 * iqr
lo = q1 - box
hi = q3 + box

enough_schools = []

for key, val in all_schools.items():
    if val >= lo and val <= hi:
        enough_schools.append(key)

len(enough_schools)
```

```python
state_dict = {
    'AS': 'American Samoa',
    'DC': 'Washington DC',
    'PR': 'Puerto Rico',
    'GU': 'Guam',
    'VI': 'Virgin Islands'
}

tuition_cost.loc[tuition_cost['state'].isnull(), 'state'] = tuition_cost.loc[tuition_cost['state'].isnull(), 'state_code'].map(state_dict)

tuition_cost.info()
```

```python
sub_cost = tuition_cost[['name', 'state', 'type', 'degree_length', 'in_state_tuition', 'out_of_state_tuition']].copy()
sub_sal = salary_potential[['rank', 'name', 'state_name', 'early_career_pay', 'mid_career_pay']].copy()
sub_div = diversity_school[['name', 'total_enrollment']].copy()

df_sub = pd.merge(sub_cost, sub_sal, how='inner', on='name')
df1 = pd.merge(df_sub, sub_div, how='inner', on='name')
```

```python
df1 = df1.drop(columns=['state_name'])
```

```python
fig = px.scatter(
    df1, x='out_of_state_tuition', y='mid_career_pay', color='state', size='total_enrollment', facet_col='type',
    labels={"in_state_tuition": "In-State Tuition", 'out_of_state_tuition': 'Out-of-State Tuition', "mid_career_pay": "Mid Career Salary", 
            'state': 'State', 'total_enrollment': 'Total Enrollment', 'type': 'Type', 'name': 'School'},
    hover_name='name'
)
fig.update_xaxes(matches=None)
fig.update_yaxes(matches=None)
fig.show()
```
